```python
from cyvcf2 import VCF
import pandas as pd
```

```python
vcf = VCF("clinvar_20260201.vcf")
variant = next(iter(vcf))  # get first variant

print(dir(variant))
```

```python
rows = []
for var in vcf:
    rows.append({
        "chrom": var.CHROM,
        "pos": var.POS,              # 1-based
        "ref": var.REF,
        "alt": ",".join(var.ALT),
        "id": var.ID,
        "qual": var.QUAL,
        "filter": var.FILTER,
        "info": dict(var.INFO),      # INFO fields as dict
        # genotype example (per-sample GT types)
        # 0=homozygous ref, 1=het, 2=homozygous alt, 3=unknown
        "gt_types": var.gt_types.tolist(),
    })

df = pd.DataFrame(rows)
df.head()
```
